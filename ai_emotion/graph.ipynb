```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
```

```python
df1 = pd.read_csv('emotionavg.csv', encoding='EUC-KR')
df = df1.head(7)
df
```

```python
import seaborn as sns

# 날짜 형식 변환
df['date'] = pd.to_datetime(df['date'])

# X축 표시를 위해 날짜와 요일을 줄바꿈으로 결합
df['x_label'] = df['date'].dt.strftime('%d') + "\n" + df['day']


# 그래프 설정
plt.figure(figsize=(10, 6))
sns.barplot(x="x_label", y="score", data=df, color="skyblue")
plt.gca().axes.get_yaxis().set_visible(False)

for i, score in enumerate(df['score']):
    plt.text(i, max(df['score']) + 15, f"{score}", ha='center', fontsize=16, color= "#797979")


plt.plot(df['x_label'], df['score'], marker='o', color='red')

for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.xlabel('')

# 그래프 출력
plt.tight_layout()
plt.show()
```
